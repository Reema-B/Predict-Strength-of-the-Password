```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data.csv',error_bad_lines=False)  # error_bad_lines ignores parser error
data.head()
```

```python
 # Password : independent variable
 # strength : dependent variable
```

```python
data['strength'].unique()

# 0 = weak
# 1 = normal
# 2 = Good
```

```python
# Only 1 missing value in password column, hence dropping the NA Value.
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
sns.countplot(data['strength'])
```

```python
pwd_data_tuple = np.array(data)
```

```python
pwd_data_tuple
```

```python
# Pass shuffled data as it will give robustness to the model.
import random
random.shuffle(pwd_data_tuple)
```

```python
# Iterate over the tupple to get the independent and dependent variables
x = [labels[0] for labels in pwd_data_tuple]   # independent variable
y = [labels[1] for labels in pwd_data_tuple]   # dependent variable
```

```python
x
```

```python
# Machine Learning only supports numeric data hence we need to convert string data to numeric data.
# TF -IDF applies importance to each word in the data
# Step-1 : Convert data to LowerCase.
# Step 2 : Compute TF for each word using formula
#        : TF = no. of occurences of a word in the doc / no. of words in the doc
# Step 3 : IDF
#        : IDF = log(Total no. of docs/no. of docs containing the words)
```

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char('WUt9IZzE0OQ7PkNE')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer = word_divide_char)
```

```python
X = vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector = X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df = pd.DataFrame(first_document_vector.T.todense(), index=vectorizer.get_feature_names(), columns = ['TF-IDF'])
```

```python
df
```

```python
df.sort_values(by=['TF-IDF'],ascending=False)
```

#### split data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

#### Apply Logistic Regression on Data as it is a classification case_study

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0, multi_class = 'multinomial')  # More then 2 classes hence multinomial
```

```python
clf.fit(X_train, y_train)  # Fit the training data.
```

#### Doing prediction for specific custom data

```python
dt = np.array(['%123@*hgy^'])
pred = vectorizer.transform(dt)   # To convert from string to numeric.
clf.predict(pred)

# Gives O/p = 2 which is Good Strength Password
```

#### Doing prediction on X_test data

```python
y_pred = clf.predict(X_test)
y_pred
```

#### checking accuracy of your model using confusion_matrix, accuracy_score

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix : \n\n', cm)
```

```python
print('Accuracy Score:', accuracy_score(y_test, y_pred))
```

#### Classification Report of the model

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
print(clf.score(X_test, y_test))
```
